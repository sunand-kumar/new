```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.pairplot(df)
```

```python
import sklearn
```

```python
from sklearn.model_selection import train_test_split
X = df[['feature1','feature2']].values
y = df['price'].values
X_train,X_test, Y_train, Y_test = train_test_split(X,y,test_size=0.3,random_state=23)
```

```python
print(X_train.shape, Y_train.shape)
```

```python
print(X_test.shape, Y_test.shape)
```

```python
import sklearn
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(4,activation = 'relu'))
model.add(Dense(4,activation = 'relu'))
model.add(Dense(4,activation = 'relu'))
model.add(Dense(1))
```

```python
model.compile(loss = 'mse', optimizer ='rmsprop', metrics = ['mse'])
```

```python
model.fit(X_train,Y_train, epochs = 200)
```

```python
model.summary()
```

```python
model.history.history
```

```python
loss = model.history.history['loss']
```

```python
sns.lineplot(x = range(len(loss)), y=loss)
plt.title("training loss per epoch")
```

```python
Y_train_pred = model.predict(X_train)
Y_test_pred = model.predict(X_test)
Y_test_pred
```

```python
Y_test_pred_df = pd.DataFrame(Y_test_pred,columns = ['pred_Y'])
Y_test_pred_df
```

```python
Y_test_df = pd.DataFrame(Y_test,columns = ['actual_Y'])
Y_test_df
```

```python
compare_df = pd.concat([Y_test_df,Y_test_pred_df],axis =1)
compare_df
```

```python
sns.scatterplot(x = 'actual_Y', y = 'pred_Y', data = compare_df)
```

```python
from sklearn.metrics import r2_score
print(r2_score(Y_train, Y_train_pred),r2_score(Y_test, Y_test_pred))
```
